# Descriptive Statistics

## Measure of Central Tendency

```python
## calculate mean, median, mode
Inc_mean = df['Income'].mean()
Inc_median = df['Income'].median()
Inc_mode = df['Income'].mode()[0]
```

```python
print('Mean:',Inc_mean)
print('Median:',Inc_median)
print('Mode:',Inc_mode)
```

```python
print('Variance:',inc_var)
print('Standard Deviation:',inc_sd)
```

# Hypothesis Testing

```python
## groupby education and income for hypothesis analysis
Edu_Inc = df.groupby(['Education'])['Income']
Edu_Inc.describe()
```

```python
Edu_Inc.mean()
```

```python
df_2nCycle = df[df['Education'] == '2n Cycle']
df_Basic = df[df['Education'] == 'Basic']
df_Graduation = df[df['Education'] == 'Graduation']
df_Master = df[df['Education'] == 'Master']
df_PhD = df[df['Education'] == 'PhD']
```

```python
## anova testing
anove_test = st.f_oneway(df_2nCycle['Income'],
                          df_Basic['Income'],
                          df_Graduation['Income'],
                          df_Master['Income'],
                          df_PhD['Income'],)
```

```python
## find pvalue
anove_test.pvalue
```

```python
if anove_test.pvalue > 0.05:
    print('Education does not affect on Income')
else:
    print('Education affects Income')
```
